```python
import pandas as pd
import os

# Get all parquet files in this dir
def get_parquet_files(directory):
    return [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

parq_files = get_parquet_files('.')
dfs = []
total = len(parq_files)
for file in parq_files:
    print(f"Reading {file} ({len(dfs) + 1}/{total})")
    df = pd.read_parquet(file)
    dfs.append(df)
```
